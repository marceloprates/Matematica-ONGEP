```python
import celluloid
import numpy as np
from matplotlib import pyplot as plt
```

```python
def draw_fraction(fraction, ax, startangle=0, step=None):
    if step is not None:
        sizes = [step]*(int(1/step))
    else:
        sizes = sorted([fraction, 1-fraction])
    
    colors = ['red']*len(sizes)
    explode = [.05]*len(sizes)
    
    cum = 0
    for i, s in enumerate(sizes):
        if cum >= fraction:
            break
        else:
            cum += s
            colors[i] = 'blue'
            explode[i] = .2
    
    ax.pie(sizes, colors=colors, explode=explode, shadow=True, startangle=startangle)
    ax.axis('equal')
    return startangle + fraction*360

fig, axes = plt.subplots(1, 5, figsize=(15, 5))
for ax in axes:
    ax.axis('off')

def draw_pipeline(subdivide):
    # Draw first fraction
    angle = draw_fraction(1/3, axes[0], step=1/6 if subdivide else None)

    # Draw plus sign
    axes[1].text(.3, .4, '+', fontsize=100)

    # Draw second fraction
    angle = draw_fraction(1/6, axes[2], angle, step=1/6 if subdivide else None)
    
    # Draw equals sign
    axes[3].text(.3, .4, '=', fontsize=100)
    
    # Draw third fraction
    draw_fraction(1/2, axes[4], step=1/6 if subdivide else None)

    camera.snap()


camera = celluloid.Camera(fig)
draw_pipeline(False)
draw_pipeline(True)
draw_pipeline(True)
animation = camera.animate(interval=800)

from IPython.display import HTML
HTML(animation.to_html5_video())
```
